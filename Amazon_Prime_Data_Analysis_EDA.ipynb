## 3. ***Data Wrangling***

### **Data Wrangling Code**

#### Step-1 :  Handling Duplicates

```python
# Identify Duplicate Values (titles.csv)

df1.duplicated().sum()
```

```python
# Identify Duplicate Values (credits.csv)

df2.duplicated().sum()
```

```python
# removing duplicate rows to ensure accuracy. (titles.csv)

df1.drop_duplicates(inplace=True)

df1.duplicated().sum()
```

```python
# removing duplicate rows to ensure accuracy. (credits.csv)

df2.drop_duplicates(inplace=True)

df2.duplicated().sum()
```

#### Step-2 : Handling Missing Values

```python
# 1. Replacing missing values with "No Description Available" in 'description' Column. (titles.csv)

df1['description'] = df1['description'].fillna("No Description Available")
```

```python
# 2. filling missing values with "Unknown" in 'age_certification' column to retain all data. (titles.csv)

df1['age_certification'] = df1['age_certification'].fillna("Unknown")
```

```python
# 3. Dropping 'seasons' Column as most values in 'seasons' column are missing because this applies only to TV shows not for movies. (titles.csv)

df1.drop(columns=['seasons'], inplace=True)
```

```python
# 4. Filling missing numerical scores with the median score. (titles.csv)

df1['imdb_score'] = df1['imdb_score'].fillna(df1['imdb_score'].median())

df1['tmdb_score'] = df1['tmdb_score'].fillna(df1['tmdb_score'].median())

df1['tmdb_popularity'] = df1['tmdb_popularity'].fillna(df1['tmdb_popularity'].median())
```

```python
# 5. Fill "null values" with 0 in 'imdb_votes' column. (titles.csv)

df1['imdb_votes'] = df1['imdb_votes'].fillna(0)
```

```python
# 6. Replacing missing 'character' column values with "Not Available". (credits.csv)

df2['character'] = df2['character'].fillna("Not Available")
```

```python
# 7. Replacing missing 'imdb_id' column values with random id value. (titles.csv)

df1['imdb_id'] = df1['imdb_id'].ffill()
```

```python
# looking for null/missing values after data wrangling process. (titles.csv)

df1.isnull().sum()
```

```python
# looking for null/missing values after data wrangling process. (credits.csv)

df2.isnull().sum()
```

#### Step-3 : Data Type Corrections (`titles.csv`)

```python
# Convert categorical columns to string
df1['title'] = df1['title'].astype(str)
df1['show_type'] = df1['show_type'].astype(str)
df1['genres'] = df1['genres'].astype(str)
df1['production_countries'] = df1['production_countries'].astype(str)

# Convert numerical columns to correct type
df1['release_year'] = df1['release_year'].astype(int)
df1['runtime'] = df1['runtime'].astype(int)
df1['imdb_votes'] = df1['imdb_votes'].astype(int) 

# Convert scores to float
df1['imdb_score'] = df1['imdb_score'].astype(float)
df1['tmdb_score'] = df1['tmdb_score'].astype(float)
df1['tmdb_popularity'] = df1['tmdb_popularity'].astype(float)
```

```python
# Dataset information after data type conversions. (titles.csv)

df1.info()
```

#### Step-4 : Normalizing Genre & Production Country Lists

```python
# Removes list formatting characters and split genres into individual words
df1['genres'] = df1['genres'].str.replace(r"[\[\]']", '', regex=True)  # Removes brackets and quotes
df1['genres'] = df1['genres'].str.replace(',', ' ')                    # Replaces commas with spaces

# Converts space-separated genres into lists, then explode into separate rows
df1['genres'] = df1['genres'].str.split()  # Splits into a list of words
df1 = df1.explode('genres')                # Expands into separate rows

df1.head()
```

```python
# Removes list formatting characters and split production countries into individual words
df1['production_countries'] = df1['production_countries'].str.replace(r"[\[\]']", '', regex=True)  # Removes brackets and quotes
df1['production_countries'] = df1['production_countries'].str.replace(',', ' ')                    # Replaces commas with spaces

# Converts space-separated genres into lists, then explode into separate rows
df1['production_countries'] = df1['production_countries'].str.split()  # Splits into a list of words
df1 = df1.explode('production_countries')                              # Expands into separate rows

df1.head()
```

#### Step-5 :  Merging `titles.csv` and `credits.csv`

```python
# Merge / joining two datasets

df = pd.merge(df1, df2, on='id', how='left')

df.info()
```

```python
# checking for duplicated rows in merged dataset.

df.duplicated().sum()
```

```python
# checking for null/missing values in merged dataset.

df.isnull().sum()
```

```python
# dropping all the null/missing values to ensure the dataset is clean.

df.dropna(inplace=True)
```

```python
df.info()
```

#### Step-6 : Saving the Cleaned Dataset

```python
# Saving cleaning datasets
df.to_csv("Cleaned_Dataset.csv", index=False)

print("Data cleaning complete. Cleaned Dataset saved.")
```
